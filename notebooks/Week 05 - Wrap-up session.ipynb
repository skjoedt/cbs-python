# CASE: Solving a peptide sequence from tandem MS spectrum

## Introduction
In a mass spectrometer, a sample is vaporized (turned into gas), and then particles from the sample are ionized. The resulting ions are placed into an electromagnetic field, which separates them based on their charge and mass. The output of the mass spectrometer is a mass spectrum, or a plot of ions' possible mass-to-charge ratio values with the intensity (actual observed frequency) of ions having these mass-to-charge values.

<center>
<img src="files/images/mass_spectrometer.jpg" width="800px" />
</center>


A special mass spectrometer setup us called a Tandem MS, or MSMS, which contains not one but *two* mass selections. In such setup the compounds in the sample are first separated by mass (shown in purple below) and in turn goes into a collision cell where the compounds collide with other molecules and are dissociated into smaller fragments (shown in green below). 

This is used in modern research to get more information about the compounds in the sample. It is often used to get information about the amino acid sequence of peptides based on what fragments are found (mass values).
<center>
<img src="files/images/MS_MS.png" width="800px" />
</center>

Here is a schematic of the locations in the protein sequence where the dissociation occurs. The resulting ions from the collisions are called b and y ions according to which half of the protein is observed. 
<center>
<img src="files/images/msms_ions.jpg" width="600px" />
</center>

The goal of this case is to infer the amino acid sequence of the peptide when given a list of masses of the resulting b ions from collisions with the full peptide. The list also contains the full peptide itself (the highest mass).

## STEP 3: Fill out the functions

```python
def load_spectrum(filename):
    """ Loads a spectrum file and return a sorted list """
    spectrum = []
    with open(filename) as f:
        line = f.readline()
        while line:
            if line[0] != "#":
                spectrum.append( float(line.rstrip()) )
            line = f.readline()
    spectrum.sort()
    return spectrum
```

```python
def load_masstable(filename):
    """ Loads mass table file and return a dictionaory """
    masstable = {}
    with open(filename) as f:
        line = f.readline()
        while line:
            if line[0] != "#":
                line = line.split()
                masstable[line[0]] = float( line[1] )
            line = f.readline()
    return masstable
```

```python
def mass_lookup(mass, masstable, tolerance=0.002):
    """ Returns the amino acid corresponding to mass """
    for aa in masstable:
        if abs(mass - masstable[aa]) <= tolerance:
            # MASS IS FOUND
            return aa
    return ""
```

```python
def ms2protein(spectrum, masstable, tolerance=0.002):
    """ Infers protein sequence from spectrum list """
    
    # PREPARE PROTEIN
    protein = ""
    
    # PREPARE MASSES
    H = 1.007825032
    O = 15.994914622
    
    for i in range(len(spectrum)):  
        if i==0:
            # FIRST MASS
            mass = spectrum[i] - H
            aa = mass_lookup(mass, masstable, tolerance)
        elif i==len(spectrum)-1:
            # LAST MASS
            mass = spectrum[i] - spectrum[i-1] - O - H
            aa = mass_lookup(mass, masstable, tolerance)
        else:
            # MIDDLE MASS
            mass = spectrum[i] - spectrum[i-1]
            aa = mass_lookup(mass, masstable, tolerance)
        if aa:
            protein += aa
    return protein
```

## STEP 4: Run it!

```python
spectrum = load_spectrum("../data/msms_bions.csv")
masstable = load_masstable("../data/monoisotopic_mass.csv")
ms2protein(spectrum, masstable)
```
