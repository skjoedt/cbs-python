## Getting the latest versions of lecture notes and exercises
Since the course is still under development lecture notes and exercises are expected to be updated during the semester. To do all this as painless as possible we should start out by setting everything up for the rest of the course.

### MAC

All you need is git. Go to http://git-scm.com/download/mac and downlaod the dmg installation file. Once downloaded, open it and open the git installer inside. Follow the instructions for installing git. 

Once installed open a terminal and navigate to the document folder where you want the course material to be downloaded, eg. 

<cb>cd /Users/christian/Documents/courses/</cb>

now clone the cbs-python github 

<cb>git clone https://github.com/skjoedt/cbs-python.git</cb>

If things worked out as expected you should have a folder called <cb>cbs-python</cb> from where you called the <cb>git clone</cb> command. This folder will store all course materials and exercises as they are updated during the semester.


#### Updating your local files
Before each lectures you should update the files to the newest versions. This is done by simply running

<cb>git checkout -f master</cb>

and 

<cb>git pull origin master</cb>

from the cbs-python directory.
