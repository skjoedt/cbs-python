You will early on be confused about python being distributed in two different versions; Python2 and Python3. You will not find alot of differences between the two versions in your python journey, but one deserves to be highlighted already at this stage: 

> Some third-party *modules* (packages) are not supported in Python3

This means that if you want to use external modules made by others, they may not be compatible with your current Python3 code. Most modules you will find describes what Python version(s) they support.

## MAC
### Python
Python comes pre-installed on Mac OS X, but due to Apple's release cycle, it's often one or even two years old. The overwhelming recommendation of the "MacPython" community is to upgrade your Python by downloading and installing a newer version from the Python standard release page.

<ol>
    <li>Download the latest Python2 release for your system to your desktop:</li>
    <ul>
        <li> [<a href="http://www.python.org/ftp/python/2.7.5/python-2.7.5-macosx10.3.dmg" target="_blank">Python2 32-bit</a> (MacOS X 10.3+)] </li>
        <li> [<a href="http://www.python.org/ftp/python/2.7.5/python-2.7.5-macosx10.6.dmg" target="_blank">Python2 64-bit</a> (MacOS X 10.6+)] </li>
    </ul>
    <li>Open a terminal: <i>/Applications/Utilities/Terminal</i></li>
    <li>Navigate to Desktop by typing <cb>cd ~/Desktop/</cb></li>
    <li>Install Python2 by running either</li>
    <cb>hdiutil mount python-2.7.5-macosx10.3.dmg</cb> for 32-bit, or<br />
    <cb>hdiutil mount python-2.7.5-macosx10.6.dmg</cb> for 64-bit.<br/>
    and then, <br />
    <cb>sudo installer -pkg "/Volumes/Python 2.7.5/Python.mpkg" -target /</cb>
</ol>
### IPython and extras
The program <i>easy_install</i> comes with python and can be used to install additional Python2 packages:

<ul>
<cb>sudo easy_install numpy</cb><br />
<cb>sudo easy_install scipy</cb><br />
<cb>sudo easy_install pandas</cb><br />
<cb>sudo easy_install tornado</cb><br />
<cb>sudo easy_install pyzmq</cb><br />
<cb>sudo easy_install pip</cb><br />
</ul>

Unfortunately easy_install does not contain the latest version of IPython that we will use in the course. Therefore use the newly installed <i>pip</i> to get it:
<cb>sudo pip install ipython</cb><br />

Now, if everything went okay you should be able to run <cb>ipython notebook</cb> and see the following window:

<img src="files/images/py_install5.jpg", width="700px">

To access the course material, you have to run the <cb>ipython notebook</cb> command from the notebook directory.

## Linux
### Python
Python comes installed with all linux distributions worth mentioning.

### IPython and extras
In Ubuntu/Debian systems run in the terminal: <br />

<cb>sudo apt-get install python-pandas python-matplotlib python-numpy python-pip python-zmq python-tornado</cb><br /><br />

In Fedora systems run in the terminal:<br />

<cb>sudo yum install python-pandas python-matplotlib python-numpy python-pip python-zmq python-tornado</cb><br /><br />

Now install IPython using pip: <br />

<cb>sudo pip install ipython</cb><br /><br />

Now check that it works by running <cb>ipython notebook</cb> in your terminal. You should see the following browser window popping up:<br />

<img src="files/images/py_install6.jpg", width="700px">

To access the course material and exercises you have to run the <cb>ipython notebook</cb> command from the notebook directory.
